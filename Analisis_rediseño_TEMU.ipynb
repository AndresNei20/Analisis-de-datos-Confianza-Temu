```python
import pandas as pd

df = pd.read_excel('Prueba Hipotesis TEMU.xlsx')
print(df)
```

```python
df_Temu = df[df['Modulo'] == 'Temu']
print(df_Temu.head)
df_Rediseño = df[df['Modulo'] == 'Rediseño']
print(df_Rediseño.head)
```

# Confianza

```python
import matplotlib.pyplot as plt
import seaborn as sns

quartiles = df.groupby('Modulo')['TRUST'].quantile([0.25, 0.75]).unstack()

plt.figure(figsize=(8, 6))
sns.boxplot(x='Modulo', y='TRUST', data=df)

for modulo in quartiles.index:
    q1 = quartiles.loc[modulo, 0.25]
    q3 = quartiles.loc[modulo, 0.75]

    plt.scatter(modulo, q1, color='red', label='Cuartil inferior (Q1)' if modulo == quartiles.index[0] else "", zorder=5)
    plt.scatter(modulo, q3, color='blue', label='Cuartil superior (Q3)' if modulo == quartiles.index[0] else "", zorder=5)

    plt.text(modulo, q1, f'{q1:.2f}', color='white', ha='center', va='bottom', fontsize=20)
    plt.text(modulo, q3, f'{q3:.2f}', color='white', ha='center', va='top', fontsize=20)

plt.title('Boxplot de TRUST por Módulo con Cuartiles y Valores')
plt.xlabel('Módulo')
plt.ylabel('TRUST')

plt.legend()

plt.show()
```

# Comparación cruzando datos de variables

Principalmente este apartado esta hecho con el fin de comparar cosas entre variables, por ejemplo como la reputación o percepción de utilidad de Temu puede afectar la seguridad que sienten sus usuarios en el **Rediseño de Temu**.

```python
df = pd.ExcelFile('DATOS CONFIANZA - TEMU.xlsx')
print(df)

sheet_names = df.sheet_names
print(sheet_names)

df_seguridad_rediseño = pd.read_excel('DATOS CONFIANZA - TEMU.xlsx', sheet_name = "Seguridad REDISEÑO" )

df_reputacion_rediseño = pd.read_excel('DATOS CONFIANZA - TEMU.xlsx', sheet_name = "Reputacion RESIDEÑO" )
```

```python
seguridad_column = df_seguridad_rediseño['¿Te sentiste seguro la primera vez que entraste a Temu? Donde 1 es muy inseguro y 10 muy seguro']
utilidad_column = df_reputacion_rediseño['  ¿Qué tan útil te parece el rediseño de Temu en una escala del 1 al 10?  ']

plt.figure(figsize=(8, 6))
plt.scatter(seguridad_column, utilidad_column)

plt.xlabel('Seguridad al entrar a Temu (1-10)')
plt.ylabel('Utilidad del rediseño de Temu (1-10)')
plt.title('Comparación de Seguridad y Utilidad del Rediseño')

plt.show()
```

```python
seguridad_column = df_seguridad_rediseño['¿Te sentiste seguro la primera vez que entraste a Temu? Donde 1 es muy inseguro y 10 muy seguro']
primera_imprecion_column = df_reputacion_rediseño['  ¿Cuál fue tu primera impresión al conocer el rediseño de Temu?  ']

plt.figure(figsize=(8, 6))
plt.scatter(seguridad_column, utilidad_column)

plt.xlabel('Seguridad al entrar a Temu (1-10)')
plt.ylabel('Primera impreción al conocer el rediseño')
plt.title('Comparación de Seguridad y Primera Impresión del Rediseño')

plt.show()
```
